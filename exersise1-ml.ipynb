```python
xmodel = np.random.rand(5,1)*20
y = 2.35 * xmodel - 3.12
y = y + np.random.normal(scale=3,size=y.shape)
y = y.reshape(-1, 1)
x=xmodel
for i in range(2,5):
    x=np.hstack((x,xmodel**i)) 
model.fit(x,y)
s6=model.score(x, y)
print(s6)
xline = np.arange(0, 20, 0.1)
xline = xline.reshape(-1, 1)
x_line=xline
for i in range(2,5):
    x_line=np.hstack((x_line,xline**i)) 
Y_line=model.predict(x_line) 
plt.xlabel('x-axis',fontname='Comic Sans MS', fontsize=17,color='aqua')
plt.ylabel('y-axis',fontname='Comic Sans MS', fontsize=17,color='aqua')
plt.title('out put',fontname='Comic Sans MS', fontsize=20,color='navy')
plt.plot(xline,Y_line,'m--',label='Part2\nk = 4 & Number = 5')
plt.plot(xmodel,y,'kP', markersize=8)
plt.grid()
plt.legend(loc=0, borderpad=1.5,facecolor='lightyellow',edgecolor='aqua')
plt.savefig('exercise1.png', dpi=300)
```
